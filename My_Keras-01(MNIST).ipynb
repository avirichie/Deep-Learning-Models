```python
from tensorflow.keras.datasets import mnist

(x_train,y_train),(x_test,y_test) =  mnist.load_data()
```

```python
print("The shape of the Training set",x_train.shape)
print("The shape of the Test set",x_test.shape)
print("The shape of the Training Labelled images",y_train.shape)
print("The shape of the Test Labelled images",y_test.shape)
```

```python
from tensorflow.keras import layers
from tensorflow.keras import models

network = models.Sequential()
network.add(layers.Dense(512,activation='relu',input_shape=(784,)))
network.add(layers.Dense(10,activation='softmax')) # maps the each input function to their respective probabilities 
```

```python
 network.compile(optimizer='rmsprop',
                 loss='categorical_crossentropy',
                 metrics=['accuracy'])
```

```python
x_train = x_train.reshape((60000,784))
x_train_norm = x_train.astype('float32')/255
print("The shape is=",x_train_norm.shape)
x_test = x_test.reshape((10000,784))
x_test_norm =  x_test.astype('float32')/255
```

```python
import matplotlib.pyplot as plt
my_img = x_train_norm[700].reshape(28,28)
plt.imshow(my_img)
```

```python
from tensorflow.keras.utils import to_categorical 

y_test= to_categorical(y_test)
y_train= to_categorical(y_train)
```

```python
network.fit(x_train_norm,y_train,epochs=5,batch_size=128)
```

```python
test_loss,test_acc = network.evaluate(x_test_norm,y_test)
print("The test accuracy is:",test_acc)
print("The test loss is: ",test_loss)
```
